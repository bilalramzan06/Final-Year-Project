```python

# Function to process a single record
def process_record(record_path, class_label, output_dir):
    file_number = os.path.basename(record_path).split('_')[
        0]  # Extract file number
    hea_file = record_path

    signals, fields = wfdb.rdsamp(hea_file)
    lead_names = fields.get(
        'sig_name', [f'Lead {i+1}' for i in range(len(signals[0]))])

    csv_file = os.path.join(output_dir, f'{file_number}_{
                            class_label}.csv')  # Include class
    with open(csv_file, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Time'] + lead_names)
        for time, data_row in zip(range(len(signals)), signals):
            csv_writer.writerow([time] + list(data_row))

    print(f'CSV file created: {csv_file}')


# Parameters
input_csv = 'D:\\Semester 7\\FYP\\preprocessing\\equal_labels.csv'
output_dir = 'D:\\Semester 7\\FYP\\preprocessing\\CSV_Data'  # Output directory
records_per_class = 1000

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Read the CSV file
with open(input_csv, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip header row
    for record_path, class_label in csv_reader:
        process_record(record_path, class_label, output_dir)
```

```python
import os
import csv


def extract_labels_to_csv(source_folder, output_file):
    """
    Extracts class labels from filenames within a source folder and saves them to a CSV file.

    Args:
      source_folder: The path to the folder containing the files.
      output_file: The path to the file where the filename and class label pairs will be written.
    """
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)

        # Write column headers
        writer.writerow(['Path', 'Class'])

        for filename in os.listdir(source_folder):
            if filename.endswith(".csv") and filename[:5].isdigit() and filename[5] == "_":
                # Extract path and class label
                file_path = os.path.join(source_folder, filename)
                class_label = filename.split('_')[1].split('.')[0]

                # Write path and class label to output file
                writer.writerow([file_path, class_label])


# Replace paths with your actual folders and desired output filename
extract_labels_to_csv(
    'D:\\Semester 7\\FYP\\preprocessing\\CSV_Data', 'output_labels.csv')
```
